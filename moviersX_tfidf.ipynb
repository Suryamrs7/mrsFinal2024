**Data Pre-Processing**

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

#ast.literal_eval(text): This line uses the ast (Abstract Syntax Trees) module's
# literal_eval function to safely evaluate the string text as a Python literal.
#It can handle literals like numbers, strings, tuples, lists, dicts, booleans, and None.

#The result of ast.literal_eval(text) is assumed to be a list of dictionaries.
#The code then iterates over each dictionary in this list.

#For each dictionary in the list, it extracts the value associated with the
#key 'name' and appends it to a list L.

#The final result is a list containing the values associated with the
# 'name' key from each dictionary in the input list.
```

```python
movies.isnull().sum()
# Check the column containing null values
```

```python
movies.dropna(subset=['overview'],inplace=True)
# Drop the rows that contains null value in overview column
```

```python
movies.isnull().sum()
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
# genres column got converted into list of genres
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(2)
# keywords column get converted into list of keywords
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
          break
    return L
# function to make list of cast containing names in them
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head(2)
# get the list of cast
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
# function to fetch director from the crew column
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head(2)
# get the list of directors
```

**Text Vectorization**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# it is used to convert the text into vectors
```

```python
cv=TfidfVectorizer(max_features=10000, stop_words='english')
# total rows = 10000
# The stop_words parameter is used to specify a list of common words that should be ignored during the vectorization process.
# In this case, you've set it to 'english', which means that common English stop words like "the," "and," "in," etc., will be excluded from the feature vectors.
# Stop words are typically removed because they often don't carry significant meaning for text analysis tasks.
```

```python
cv
```

```python
vector= cv.fit_transform(new_movies['tags'].values.astype('U')).toarray()
# This code employs CountVectorizer to transform text 'tags' data from a DataFrame into a matrix
# where rows correspond to tags,
# columns represent unique words, and each cell holds the count of words in the respective tag.
```

**Similarity Calculation**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
from fuzzywuzzy import process

def movie_finder(title, all_titles):
    closest_match = process.extractOne(title, all_titles)
    return closest_match[0]

def recommend(movies):
    # Use movie_finder to correct the user input
    corrected_movie = movie_finder(movies, new_movies['title'].tolist())

    # Print the corrected movie name
    print(f"Corrected Movie Input: {corrected_movie}")

    # Get the index of the corrected movie in the dataset
    index = new_movies[new_movies['title'] == corrected_movie].index[0]

    # Calculate similarity scores
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])

    # Print top 5 recommendations
    print("\nTop 5 Recommendations:")
    for i in distance[1:21]:
        print(new_movies.iloc[i[0]].title)
        # print(new_movies.iloc[i[0]].popularity)
        print(f"Similarity Score: {[i[1]]}")

# Example usage
user_input = "the dark knight rises"
recommend(user_input)
```
